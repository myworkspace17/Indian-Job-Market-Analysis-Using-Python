# What is the correlation between educational qualifications and salary ranges?


```python
# Group the data by 'Education Requirement' and 'Salary Range' and count the number of job postings
edu_salary_counts = df.groupby(['Education Requirement', 'Salary Range']).size().reset_index(name='Number of Postings')

# Create a pivot table to format the data for visualization
edu_salary_pivot = edu_salary_counts.pivot(index='Education Requirement', columns='Salary Range', values='Number of Postings')

# Plot a heatmap to visualize the correlation between education requirement and salary range
plt.figure(figsize=(12, 8))
sns.heatmap(edu_salary_pivot, annot=True, cmap='Blues', fmt='d', linewidths=0.5)

plt.title('Correlation between Education Requirement and Salary Range', fontsize=16)
plt.xlabel('Salary Range', fontsize=12)
plt.ylabel('Education Requirement', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Group the data by 'Education Requirement' and 'Salary Range' and count the number of job postings
edu_salary_counts = df.groupby(['Education Requirement', 'Salary Range']).size().reset_index(name='Number of Postings')

# Create a histogram-like effect using FacetGrid
plt.figure(figsize=(12, 8))
g = sns.FacetGrid(edu_salary_counts, col='Education Requirement', col_wrap=3, height=4)
g.map_dataframe(sns.histplot, x='Salary Range', weights='Number of Postings', bins=len(edu_salary_counts['Salary Range'].unique()), kde=False)
g.set_axis_labels('Salary Range', 'Number of Postings')
g.set_titles(col_template="{col_name}")
g.add_legend()

plt.subplots_adjust(top=0.9)
g.fig.suptitle('Distribution of Job Postings by Education Requirement and Salary Range', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a contingency table between Education Requirement and Job Title
edu_job_crosstab = pd.crosstab(df['Education Requirement'], df['Job Title'])

# Normalize the table to get percentages
edu_job_crosstab_percentage = edu_job_crosstab.div(edu_job_crosstab.sum(axis=1), axis=0) * 100

# Create a heatmap to visualize the correlation
plt.figure(figsize=(12, 8))
sns.heatmap(edu_job_crosstab_percentage, annot=True, fmt='.1f', cmap='coolwarm', cbar_kws={'label': 'Percentage (%)'})

# Add title
plt.title('Correlation between Education Requirement and Job Titles', fontsize=16)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Education Requirement', fontsize=12)

# Show the plot
plt.show()
```

# How does the number of job postings vary by month? (Line chart of postings over time)

# What is the average duration between job posting and application deadline? (Histogram)


```python
df
```

```python
# Ensure 'Posted Date' and 'Application Deadline' are datetime objects
df['Posted Date'] = pd.to_datetime(df['Posted Date'], errors='coerce')
df['Application Deadline'] = pd.to_datetime(df['Application Deadline'], errors='coerce')

# Calculate the duration between job posting and application deadline
df['Duration (Days)'] = (df['Application Deadline'] - df['Posted Date']).dt.days

# Drop any rows where the duration is NaN (in case of missing or invalid dates)
df = df.dropna(subset=['Duration (Days)'])

# Create a histogram to visualize the distribution of the duration
plt.figure(figsize=(10, 6))
plt.hist(df['Duration (Days)'], bins=30, color='green', edgecolor='black')

# Adding labels and title
plt.xlabel('Duration (Days)')
plt.ylabel('Frequency')
plt.title('Distribution of Duration Between Job Posting and Application Deadline')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Ensure 'Posted Date' and 'Application Deadline' are datetime objects
df['Posted Date'] = pd.to_datetime(df['Posted Date'], errors='coerce')
df['Application Deadline'] = pd.to_datetime(df['Application Deadline'], errors='coerce')

# Calculate the duration between job posting and application deadline
df['Duration (Days)'] = (df['Application Deadline'] - df['Posted Date']).dt.days

# Drop any rows where the duration is NaN (in case of missing or invalid dates)
df = df.dropna(subset=['Duration (Days)'])

# Create a figure for the plot
plt.figure(figsize=(12, 6))

# Create a histogram with a KDE overlay
sns.histplot(df['Duration (Days)'], bins=30, kde=True, color='skyblue', edgecolor='black', stat='density')

# Adding labels and title
plt.xlabel('Duration (Days)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.title('Distribution of Duration Between Job Posting and Application Deadline', fontsize=16)

# Display the plot
plt.tight_layout()
plt.show()
```

# Which companies offer the most remote work opportunities?

```python
import matplotlib.pyplot as plt

# Filter the data for remote job postings
remote_jobs = df[df['Remote/Onsite'] == 'Remote']

# Display the filtered 'Company Name' for remote jobs
remote_companies = remote_jobs['Company Name']
print("Companies offering remote jobs:")
print(remote_companies)

# Group by 'Company Name' and count the number of remote jobs per company
remote_jobs_count = remote_jobs.groupby('Company Name').size()

# Display the result sorted by the number of remote job postings
print("\nCount of remote job postings by company:")
print(remote_jobs_count.sort_values(ascending=False))

# Plotting the distribution of remote jobs per company (top 10 companies)
plt.figure(figsize=(12, 6))
remote_jobs_count.sort_values(ascending=False).head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Companies Offering Remote Work Opportunities')
plt.xlabel('Company Name')
plt.ylabel('Number of Remote Job Postings')
plt.xticks(rotation=45, ha='right')  # Rotate company names for better readability
plt.tight_layout()
plt.show()
```
